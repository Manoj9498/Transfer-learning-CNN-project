```python
import os
import shutil
import glob
from tqdm import tqdm
```

```python
Raw_Dir=r'C:\Users\manoj\Downloads\mrlEyes_2018_01'
for dirpath, dirname, filenames in os.walk(Raw_Dir):
    for i in tqdm([f for f in filenames if f.endswith('.png')]):
        if i.split ('_')[4]=='0':
            shutil.copy(src=dirpath+'/'+i, dst=r'C:\Users\manoj\OneDrive\Desktop\namma project\close_eyes')
            
        elif i.split ('_')[4]=='1':
            shutil.copy(src=dirpath+'/'+i, dst=r'C:\Users\manoj\OneDrive\Desktop\namma project\open_eyes')
```

```python
Raw_Dir=r'C:\Users\manoj\OneDrive\Desktop\namma project\open_eyes'
for dirpath, dirname, filenames in os.walk(Raw_Dir):
    z=0
    for i in tqdm([f for f in filenames if f.endswith('.png')]):
        shutil.move(src=dirpath+'/'+i, dst=r'C:\Users\manoj\OneDrive\Desktop\namma project\test_data\open_eyes')
        z=z+1
        if z==8000:
            break
            
```
